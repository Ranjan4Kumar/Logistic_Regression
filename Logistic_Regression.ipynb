## Logistic Regression :-
It is also a ML models .
It is used for binary classification tasks, it involves prediction one of two possible outcomes or classess.
It is also a type of regression analysis that is particularly well-suited for problems where the dependent variable is categorical and takes on binary values.( yes or no, true or false, o or 1)

### How actually logistic regression works
- 1. Sigmoid Fucntion(Logistic Fucntion) : the model whihc uses the logistic fucntion (also called sigmoid fucntion) to transform a linear combination of input features into a values between 0 and 1.




```python
# Graphical representation of sigmoid fucntion
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(x):
  return 1/(1+np.exp(-x))
# generating x values
x = np.linspace(-7,7,200)
# calculating corresponding y values using the sigmoid
y = sigmoid(x)

plt.plot(x,y, label = 'Sigmoid Function', linewidth = 2)
plt.grid(True)
plt.legend()
plt.show()
```

- 2. Linear Combination : The logistic regression assumes alinear relatioship between the input features and the log-odds of the output. The linear combination z is calculated as follows
z = b0 + b1.x1 + b2.x2 + b3.x3 + ... + bn.xn
b0 -> intecept term
b1,b2,b3,...,bn are the corfficients for each inout features (x1,x2,x3,....,xn)

- 3. Predicting Probabilities : Once you have z, you can pass through the sigmoid fucntion to get the probability f the (+) class (class1).

P(Y = 1) = sigma(z) = 1/(1+e^-z)
Similarly, the probabilities of the negative class (class 0)is
P(Y = 0) = 1 - P(Y=1)

- 4. Thresholding :  To make binary prediction, you can typically choose a threshold (e.g 0.5) and classify data points with predicted probabilites greater the the threshold as belogning to class 1 and those below the threshold as belogning to class 0.

#### Precision Recall Tradeoff:
When your data is perfectly seprable, you will achieve a value of 1.0 for precision and recall.
In most practical situations, this is impossible to achieve, because of this condition tradeoff occurr.

- If we increase a precision and the recall will decrease anf if we increase recall prcision will decrese. We seek for an optimal condtion that is called optimal threshold. This optimal threshold lead you to an optimal tradeoff.

- Optimizing the precision/recall tradeoff comes down to finding an optimal threshold by looking at the precision and recall.
The easiest way to be sure that your set your balance right is the F1 score.

F1 = 2[(Recall*Precision)]/(Recall + precision)

- F1 score makes sure that you achieve a good balance between precision and recall.
- A high value of F1 score is a good indicator that your model performs well.

- F1 score = 0: It indicates the worst possible performance,It means the model fails to correclty classify any positive instances.

- F1 score = 1: This represent perfect performance, where the model achieves both perfect precision and recall.
- F1 score between 0 and 1: F1 score between 0 and 1 indicates various degress of model performance, A higher F1 score indicates better overall performance, with the balance between precision and recall.

### Roc (Reciver operator Curve):
It display the realtionship between sensitivity and fall-out(FP/TN+FP).
This curve allows us to find good probability threshold.
Threshold are the deicision point used by the midel for classification.

ROC curve is the curve betwee n( True positive rate and False positive rate)

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_curve, roc_auc_score, auc

# Load the data
iris = datasets.load_iris()
X = iris.data
y = iris.target

# we will convetr this into binary classification problems
# class 0(setosa), class 1(Versicolr + verginica)
y_binary = (y==0).astype(int)

# Split the dataset into training and testing sets
X_train,X_test,y_train,y_test = train_test_split(X,y_binary, test_size = 0.3,random_state = 42)

#

model1 = LogisticRegression(solver = 'liblinear')
model1.fit(X_train,y_train)
```

```python
# We will fit out model
y_prob = model1.predict_proba(X_test)[:,1]

# Calculate the ROC
fpr,tpr, thresholds = roc_curve(y_test,y_prob)

# Calculate AUC (Area under the curve)
roc_auc = auc(fpr, tpr)

plt.plot(fpr,tpr, color = 'darkorange', lw = 2, label = 'ROC Curve (AUC = {:.2f})'.format(roc_auc))

plt.plot([0,1],[0,1], color = 'navy', lw = 2, linestyle = '--')

plt.xlim([0.0,1.0])
plt.ylim([0.0,1.05])
plt.legend(loc = 'lower right')
plt.show()
```
